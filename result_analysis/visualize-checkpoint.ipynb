## 1 Slow Down

```python
image_scale = [1, 4, 9, 16]
n_gpu_time = [0.0011,0.0019,0.0034,0.0055]
n_openmp_time = [0.079566,
0.191321,
0.411753,
0.724317
]
n_serial_time = [1.660690,6.626279,14.901165,26.465120]

fig, ax = plt.subplots(figsize=(6,6))
ax.set_xscale("log")
ax.set_yscale("log")
ax.plot(image_scale, n_gpu_time, 'bo-', label=r'GPU')
ax.plot(image_scale, n_openmp_time, 'co-', label=r'OpenMP')
ax.plot(image_scale, n_serial_time, 'go-', label=r'Serial $O(n)$')

ax.set_xlabel("Problem Size(Image Sizes)")
ax.set_ylabel("Time(s)")

ax.set_title("Problem Size vs Time(s) in log coordinate")
plt.grid()
ax.legend()
```

```python
# slow down calculation
openmp_index = (np.log(0.724317) - np.log(0.079566)) / (np.log(16) - np.log(1)) - 1
gpu_index = (np.log(0.0055) - np.log(0.0011)) / (np.log(16) - np.log(1)) - 1
serial_index = (np.log(26.384604) - np.log(1.658905)) / (np.log(16) - np.log(1)) - 1
print(openmp_index,
gpu_index,
serial_index)
```

```python
rgba2gray = np.array([0.010936, 0.014840,  0.018960, 0.025472])
post_pro_with_max = np.array([0.008574, 0.032461, 0.072081, 0.128096])
find_max = np.array([0.002182, 0.009086, 0.021112, 0.037780])
conv3 = np.array([0.054521, 0.122779, 0.272671, 0.485060])
gray2rgba = np.array([ 0.005057, 0.021085, 0.047885,  0.085534])

post_pro = post_pro_with_max - find_max
post_pro

import pandas as pd
import matplotlib.pyplot as plt

Problem_size=["image","image 4x","image 9x","image 16x"]
data={
    "Conversion(RGBA to Grayscale)":rgba2gray,
    "Edge Detection":conv3,
    "Get Globel Max":find_max,
    "Grayscale Stretch":post_pro,
    "Conversion(Grayscale to RGBA)":gray2rgba,
}

df=pd.DataFrame(data,index=Problem_size)

df.plot(kind="bar",stacked=True,figsize=(5,5))
plt.legend(loc="lower left",bbox_to_anchor=(0,1.0))
plt.xlabel("Image Size")
plt.ylabel("runtime(s) of each part")
plt.show()
data
```

```python
# Get ratio
all_time = rgba2gray + post_pro + find_max + conv3 + gray2rgba
rgba2gray = rgba2gray / all_time
post_pro = post_pro / all_time
find_max = find_max / all_time
conv3 = conv3 / all_time
gray2rgba = gray2rgba / all_time

import pandas as pd
import matplotlib.pyplot as plt

Problem_size=["image","image 4x","image 9x","image 16x"]
data={
    "Conversion(RGBA to Grayscale)":rgba2gray,
    "Edge Detection":conv3,
    "Get Globel Max":find_max,
    "Grayscale Stretch":post_pro,
    "Conversion(Grayscale to RGBA)":gray2rgba,
}

df=pd.DataFrame(data,index=Problem_size)

df.plot(kind="bar",stacked=True,figsize=(5,5))
plt.legend(loc="lower left",bbox_to_anchor=(0,1.0))
plt.xlabel("Image Size")
plt.ylabel("runtime percentage of each part")
plt.show()
data
```

## 2 Strong sclaing

```python
import numpy as np
threads_num = np.array([4,8,16,32,64])
strong_time = np.array([7.332812,
3.966220,
2.101034,
1.163144,
0.757214])
serial = 26.3846

eff = serial / (threads_num * strong_time) * 100
eff
```

```python
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(['4','8','16','32','64'], strong_time, marker='o', color='royalblue', label='OpenMP')
ax.plot(['4','8','16','32','64'], [7.3332, 3.666, 1.833, 0.9165, 0.4582], 'g--', label='Reference(Slope=-1)')
ax.set_xlabel("Threads")
ax.set_ylabel("Runtime (sec)")

ax.set_title("log-log-scale Strong Scaling(OpenMP)")
ax.set_yscale('log')
plt.grid()
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(['4','8','16','32','64'], eff, marker='o', color='royalblue')
ax.set_xlabel("Threads")
ax.set_ylabel("Percentage (%)")

ax.set_title("log-log-scale Strong Scaling Efficiency (OpenMP)")
plt.grid()
ax.legend()
```
